### Regex

#### Regex-1: Identifying word tokens from the text & extracting those tokens

```python
import regex as re

def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

```python
testtext = "Let's defeat SARS-CoV-2 together in 2020! www.cdc.gov"
testtokens = tokenize(testtext)
 
print(testtokens)
```

#### Importing natural language toolkit and downloading some stopwords in english

```python
import nltk 
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

#### Removing Stop words from Token list

```python
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]
cleantokens= remove_stop(testtokens)
print(cleantokens)
```

#### Streamlining the process of converting text to lower, identifying tokens & cleaning them by removing stopwords

```python
pipeline = [str.lower, tokenize, remove_stop]

def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

#### Creating tokens column to hold list of tokens

```python
df['tokens'] = df['text'].apply(prepare, pipeline=pipeline)
```

#### New column to hold number of tokens

```python
df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

#### Identifying frequency of tokens using counter class

```python
from collections import Counter

tokens = tokenize("She likes my cats and my cats like my sofa.")

testcount= Counter(tokens)
print(testcount)
```

#### Updating the counter with new list of tokens

```python
more_tokens = tokenize("She likes dogs and cats.")
testcount.update(more_tokens)
print(testcount)
```

#### Identifying top 5 most commonly used word in our dataset

```python
counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

#### Creating a dataset to hold tokens with frequency more than one

```python
freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

#### Sorting dataset with freq in descending order

```python
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

#### Stats for freq_df dataset

```python
freq_df.describe()
```
