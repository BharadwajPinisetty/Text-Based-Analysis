# Sentiment Analysis

#### imports the pandas library and then reads the csv file

```python
import pandas as pd 
df = pd.read_csv("C:\Amazon-Product-Reviews - Amazon Product Review (1).csv")
df.sample(10)
```

#### provides with some basic information about the Opinion Lexicon, including the total number of words and examples of positive and negative words.

```python
from sklearn import preprocessing
import nltk
nltk.download('opinion_lexicon')
from nltk.corpus import opinion_lexicon
from nltk.tokenize import word_tokenize

print('Total number of words in opinion lexicon', len(opinion_lexicon.words()))
print('Examples of positive words in opinion lexicon',
      opinion_lexicon.positive()[:10])
print('Examples of negative words in opinion lexicon',
      opinion_lexicon.negative()[:10])
```

#### sets up a dictionary where each word from the Opinion Lexicon is paired with its sentiment score

```python
# Let's create a dictionary which we can use for scoring our review text
nltk.download('punkt')
pos_score = 1
neg_score = -1
word_dict = {}
 
# Adding the positive words to the dictionary
for word in opinion_lexicon.positive():
        word_dict[word] = pos_score
      
# Adding the negative words to the dictionary
for word in opinion_lexicon.negative():
        word_dict[word] = neg_score
```

#### bing_liu_score function aims to calculate the sentiment score of a given text using the Bing Liu lexicon

```python
def bing_liu_score(text):
    sentiment_score = 0
    bag_of_words = word_tokenize(text.lower())
    for word in bag_of_words:
        if word in word_dict:
            sentiment_score += word_dict[word]
    return sentiment_score
```

# Calculating the Bing_Liu_Score

```python
# Fill missing values in 'text' column with 'no review'
df['review_body'].fillna('no review', inplace=True)

# Apply bing_liu_score function to each text in 'text' column and store results in 'Bing_Liu_Score' column
df['Bing_Liu_Score'] = df['review_body'].apply(bing_liu_score)
```

#### Prints star_rating, review_body, and the corresponding Bing Liu sentiment score for each review¶

```python
df[['star_rating',"review_body", 'Bing_Liu_Score']].head(10)
```

#### groups the DataFrame df by the 'star_rating' column and calculates the mean of the 'Bing_Liu_Score'

```python
df.groupby('star_rating').agg({'Bing_Liu_Score':'mean'})
```
